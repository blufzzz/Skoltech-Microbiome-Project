```python
import pandas as pd
import glob
```

```python
# folders are the databases folders, there should be a folder for each database in use in the current repo directory

folders = glob.glob('???')
pivots_o = []
pivots_f = []
pivots_g = []
```

```python
# pivots are the files for normalized csv data that were generated before

for f in folders :
    pivots_o += glob.glob(f + '/' + 'pivot_o_normalized.csv')
    pivots_f += glob.glob(f + '/' + 'pivot_f_normalized.csv')
    pivots_g += glob.glob(f + '/' + 'pivot_g_normalized.csv')
```

```python
# displaying list of normalized pivot tables from all databases directories

pivots_g
```

```python
frames_o = []
frames_f = []
frames_g = []
```

```python
# adding all the normalized data to a list of frames for each taxonomic level (one database for each list node)

for p in pivots_o :
    frames_o.append(pd.read_csv(p, sep = ';'))
for p in pivots_f :
    frames_f.append(pd.read_csv(p, sep = ';'))
for p in pivots_g :
    frames_g.append(pd.read_csv(p, sep = ';'))
```

```python
frames_g[3]
```

```python
# merging all the lists to single dataframes
# columns are the union of columns of the frames from each database, sorted in alphabetical order
# note that now rows are also sorted alphabetically, so some of the databases are not stored sequentially

key= frames_o[0].columns[0]
result_o = frames_o[0]
for df_o in frames_o[1:4] :
    result_o = pd.merge(result_o, df_o, how = 'outer', sort=True)
    
key= frames_f[0].columns[0]
result_f = frames_f[0]
for df_f in frames_f[1:4] :
    result_f = pd.merge(result_f, df_f, how = 'outer', sort=True)
    
key= frames_g[0].columns[0]
result_g = frames_g[0]
for df_g in frames_g[1:4] :
    result_g = pd.merge(result_g, df_g, how = 'outer', sort=True)
```

```python
result_g
```

```python
# filling the missing data with zeros (replacing None values created when new columns were added  by 0)

result_o.fillna(0, inplace = True)
result_f.fillna(0, inplace = True)
result_g.fillna(0, inplace = True)
```

```python
# storaging the merged dataframes to the csv tables

result_o.to_csv('pivot_o.csv', sep=';', index=False)
result_f.to_csv('pivot_f.csv', sep=';', index=False)
result_g.to_csv('pivot_g.csv', sep=';', index=False)
```
