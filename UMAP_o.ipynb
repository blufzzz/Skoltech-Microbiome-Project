```python
# paths to the isomap projected database and the pivot table (to take the masks for coloring databases)

paths = [
    "./isomap_o_dim5.csv",
    "./pivot_o.csv"
]
```

```python
# load dataset with isomap projected data, pca projected is not recommended beacuse of the high computational cost 

dataset = np.genfromtxt(paths[0], delimiter=';')
```

```python
dataset.shape
```

```python
# add databases names to convert them into colors

color_info = pd.read_csv(paths[1], delimiter = ';', low_memory=False)
```

```python
color_info
```

```python
# insert column for colors

color_info.insert(1, "color", True)
```

```python
# copy the names to that column

color_info["color"] = color_info["Unnamed: 0"]
```

```python
# function to convert names to colors. Suffix information is enough. Each database has specific color

def check(name) :
    if (name.startswith("103")) :
        return 'b' # 'AGP'
    elif (name.startswith("EP")) :
        return 'r' # 'PTB'
    elif (name.startswith("HMP2_") or (name.startswith("otu"))) :
        return 'g' # 'T2D'
    else :
        return 'y' # 'IBD'
```

```python
# apply coloring to the full dataset

color_info["color"] = color_info["color"].apply(check)
```

```python
# Check that the databases are marked properly by the sizes of colored fragments

color_info["color"].value_counts()
```

## Run the fragments below as many times as many embeddings needed
#### Before each plot list of parameters is in comments
#### Calculate the embedding, store it to a specific file (don't forget to rename it first) and run the specified plot for the result

```python
# construct the embedding, for the alternative choice of parameters refer to guide on UMAP, i.e. in report
# to recalculate the previous choices of parameters refer to the plots below

embedding = umap.UMAP(n_neighbors=10, n_components=2, min_dist=0.0, spread = 1.5, metric='euclidean').fit(dataset)
```

```python
embedding.embedding_
```

```python
# store the embedding to the csv table. Rename the file each time a new embedding is generated

np.savetxt("umap_o_dim2_nn20.csv", embedding.embedding_, delimiter=";")
```
